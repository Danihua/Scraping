```python
# Importacion de herraminetas necesarias
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://dbs.cl/capilar/shampoo'

page = requests.get(url)
page = BeautifulSoup(page.content, 'html.parser')
print(f'page: \n{page.prettify()}')
```

```python
page.status_code
```

```python
all_page = page.find_all(class_='products list items product-items')
print(len(all_page))
for product in all_page:
    name_product = product.find_all(class_="product name product-item-name")
    print(str(name_product))

    name_product = product.find_all(class_="price")
    print(name_product)
    price_products = page.find(class_="price")
    print(int(price_products.string.replace("$","").replace(".","")))
```
